###### hist(데이터 리스트) - 

```python
import matplotlib.pyplot as plt
import random

# plt.hist([1, 1, 2, 3, 4, 5])
# 주사위 던지기
dice = []
for i in range(100):
    dice.append(random.randint(1,6))
print(dice)

plt.hist(dice, bins=6) # 가로축의 구간의 개수
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

# with open("datas/temp_sroul.cvs", "r", encoding= "") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)
    next(data) #: 헤더 없이 출력

    temp_high = []
    for row in data:
        if row[-1] != '':
            row[0] = row[0].replace('\t', '')   # 년도 탭(공백) 제거
            if row[0].split('-')[1] == '08':
                    temp_high.append(float(row[-1]))

print(len(temp_high))
# print(temp_high)
plt.rc('font', family='Malgun Gothic')
plt.title("서울의 8월 최고기온 분포")
plt.hist(temp_high, bins=100, color='r')
plt.xlabel('온도')
plt.ylabel('빈도수')
plt.show()
```
