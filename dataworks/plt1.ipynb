### 시각화 라이브러리 사용 - matplotlib

```python
import matplotlib.pyplot as plt

plt.title("Basic Graph")
# plt.plot([1, 2, 3, 4]) # x축값이 생략되면 y축 값만 사용된다.
plt.plot([1, 2, 3, 4], [10, 40, 20, 30]) # 가로(x)축 값, 세로(Y)축 값 (1,10), (2,40)
plt.show()
```

```python
# 변수를 사용한 코드
import matplotlib.pyplot as plt

plt.title("Basic Graph")
# plt.plot([1, 2, 3, 4]) # x축값이 생략되면 y축 값만 사용된다.
x = [1, 2, 3, 4]
y = [10, 40, 20, 30] 
plt.plot(x, y, color='k') # color='k'에서 k는 black, 원래 색상은 앞글자로 표기하지만 블랙만 마지막 글자
plt.show()
```

```python
import matplotlib.pyplot as plt

# 한글 오류 해결
plt.rc('font', family='Malgun Gothic')
# pad(패딩) - 여백, 10-> 10px, loc(location) - Left, right, center
plt.title("시험 성적 그래프", pad=10, loc="right")     
plt.plot([1, 2, 3, 4], [90, 40, 85, 70]) # 가로(x)축 값, 세로(Y)축 값 (1,10), (2,40)
plt.xlabel("student number") # x축 제목
plt.ylabel("score")          # y축 제목
# 격자 형태
axis = {'both', 'x','y'}
# plt.grid()                 # default - 가로, 세로 표시
plt.grid(True, axis='y')     # axis='y' - y축 표시선 출력
plt.show()
```

```python
import matplotlib.pyplot as plt

# 한글 오류 해결
plt.rc('font', family='Malgun Gothic')
plt.title("범례 및 색상 표시")
plt.plot([10, 20, 30, 40], label="비만율", color='r') # r - red, g - green
plt.plot([40, 30, 20, 10], label="출생율", color='g')
plt.legend() # 범례(소제목) 필수 명시
plt.show()
```

```python
import matplotlib.pyplot as plt

# 한글 오류 해결
plt.rc('font', family='Malgun Gothic')
plt.title("마크 표시")
plt.plot([10, 20, 30, 40], 'b.', label="circle") # r - red, g - green
plt.plot([40, 30, 20, 10], 'r^', label="triangle")
plt.legend() # 범례(소제목) 필수 명시
plt.show()
```

```python
# 막대 그래프(세로 막대)
import matplotlib.pyplot as plt

# x = range(6)
print(x)
y = [1, 2, 3, 5, 6, 7]

plt.bar(range(6), y)
plt.show()
```

```python
# 막대 그래프(세로 막대)
import matplotlib.pyplot as plt

x = range(3)
y = [100, 400, 900]

# 세로 막대 그래프
plt.bar(range(3), y, color=['r','g','b'], width=0.6) # width-막대너비

# 눈금 표시
plt.xticks(x, ['2018', '2019','2020'])
plt.show()
```

```python
# 가로막대(수평 막대) 그래프
import matplotlib.pyplot as plt

y = range(3)
x = [100, 400, 900]

# 세로 막대 그래프
plt.barh(y, x, color=['r','g','b'], height=0.6) # width-막대너비

# 눈금 표시
plt.yticks(y, ['2018', '2019','2020'])
plt.show()
```
