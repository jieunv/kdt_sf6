```python
# 최고 기온만 리스트에 담기
import csv
import matplotlib.pyplot as plt

# with open("datas/temp_sroul.cvs", "r", encoding= "") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)
    # print(data)
    # # 헤더 - next(data) : 다음 행으로 이동
    # header = next(data) : 헤더만 출력
    # print(header)
    next(data) #: 헤더 없이 출력

    temp_high = []
    for row in data:
        if row[-1] != '':
            row[0] = row[0].replace('\t', '')   # 년도 탭(공백) 제거
            if int(row[0].split('-')[0]) >= 1980: # 연도를 정수화 비교
                temp_high.append(float(row[-1]))

print(len(temp_high))
# print(temp_high)
plt.rc('font', family='Malgun Gothic')
plt.title("1980년 이후 서울의 최고기온")
plt.plot(temp_high, color='r')
plt.show()
```

```python
# 최고 기온(최대값) 계산하기

# with open("datas/temp_sroul.cvs", "r", encoding= "") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)
    next(data) #: 헤더 없이 출력

    # for row in data:
    #     if row[-1] != '':
    #         row[-1] = float(row[-1])
            
    max_temp = -999  #최대값 설정
    max_date = ""    # 최고기온 날짜 
    # 날짜 - row[0]
    for row in data:
        if row[-1] != "":
            row[-1] = float(row[-1])
            row[0] = row[0].replace('\t', '')   # '\t'를 제거
            if row[-1] > max_temp:
                max_temp = row[-1]
                max_date = row[0]
    # print(f'최고기온: {max_temp}, 날짜: {max_date}')
    # print(max_date.strip(), max_temp)
    print(f'최고기온: {max_temp}, 날짜: {max_date}')
            
# print(max(temp_high)) : 최대값을 구할 수 있지만 날짜를 모름
```
